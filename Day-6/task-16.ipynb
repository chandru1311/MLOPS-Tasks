```python
import pandas as pd
df = pd.read_csv("civic_issue_dataset.csv")
df.head()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
model = MultinomialNB()
v = CountVectorizer()
features = ['Issue_Type','Department','Zone','Has_Photo','Has_Voice_Note','Hour_of_Day','Num_Upvotes','Population_Density','Dept_Current_Workload','Weather_Severity_Index']
X = df[features]
y = df['Severity_Label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)
X_train_text = X_train.astype(str).agg(' '.join, axis=1)
X_test_text = X_test.astype(str).agg(' '.join, axis=1)
X_train_vec = v.fit_transform(X_train_text)
X_test_vec = v.transform(X_test_text)
model.fit(X_train_vec, y_train)
print('Test score:', model.score(X_test_vec, y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=10)
model_knn.fit(X_train_vec.toarray(), y_train)
print('KNN score:', model_knn.score(X_test_vec.toarray(), y_test))
```
