```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('civic_issue_dataset.csv')

X2 = df[['Severity_Score', 'Num_Upvotes', 'Has_Photo', 'Has_Voice_Note', 'Population_Density', 'Weather_Severity_Index']]
y2 = df['Priority_Score']

model2 = LinearRegression()
model2.fit(X2, y2)

print("\nCoefficients:", model2.coef_)
print("Intercept:", model2.intercept_)
print("R² Score:", model2.score(X2, y2))


```
